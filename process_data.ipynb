```python
import pandas as pd
import glob
import os
# Load data
folder = 'gen_data'
# get all csv from this folder except 'customers.csv'
pattern = os.path.join(folder, '*.csv')
files = glob.glob(pattern)
files = [f for f in files if not f.endswith('customers.csv')]
# read all csv files and concatenate them into a single DataFrame
dataframes = [pd.read_csv(f, sep='|') for f in files]
df = pd.concat(dataframes, ignore_index=True)
```

```python
'''
Columns:
['ssn', 'cc_num', 'first', 'last', 'gender', 'street', 'city', 'state',
       'zip', 'lat', 'long', 'city_pop', 'job', 'dob', 'acct_num', 'profile',
       'trans_num', 'trans_date', 'trans_time', 'unix_time', 'category', 'amt',
       'is_fraud', 'merchant', 'merch_lat', 'merch_long']'''
df['city_pop'] = df['city_pop'].astype(int)
df['acct_num'] = df['acct_num'].astype(int)
# Remove trans_date and trans_time
df = df.drop(columns=['trans_date','trans_time'])
df['unix_time'] = df['unix_time'].astype(int)
```

```python
df['dob_day'] = pd.to_datetime(df['dob']).apply(lambda x: x.day)
df['dob_month'] = pd.to_datetime(df['dob']).apply(lambda x: x.month)
df['dob_year'] = pd.to_datetime(df['dob']).apply(lambda x: x.year)
df = df.drop(columns=['dob'])
```

```python
# Remove profile
df = df.drop(columns=['profile'])
```

```python
y = df['is_fraud']
df = df.drop(columns=['is_fraud'])
```

```python
# Categorize if columns are numeric or categorical
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
print(f'Numeric columns: {numeric_cols}')
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
print(f'Categorical columns: {categorical_cols}')
```

```python
# SSN, CC_NUM, first, last, street, city, zip, job, trans_num and merchant should be encoded as label encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for col in ['ssn', 'cc_num', 'first', 'last', 'street', 'city', 'zip', 'job',
            'trans_num', 'merchant']:
    df[col] = label_encoder.fit_transform(df[col].astype(str))
    
```

```python
# use OneHotEncoder to encode other categorical columns: 'gender', 'state', 'category'
from sklearn.preprocessing import OneHotEncoder
one_hot_cols = []
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded_cols = encoder.fit_transform(df[categorical_cols])
# Create a DataFrame with the encoded columns
encoded_df = pd.DataFrame(encoded_cols, columns=encoder.get_feature_names_out(categorical_cols))
encoded_df
```
